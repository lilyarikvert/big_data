```python
"""
Этап 2
Создайте столбец source_type для датафрейма log из материалов занятия по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic;
для источников paid и email из России — ставится ad;
для источников paid и email не из России — ставится other;
все остальные варианты берутся из traffic_source без изменений.
"""
log = pd.read_csv('visit_log.csv', sep = ';')
log
```

```python
log = pd.read_csv('visit_log.csv', sep = ';')
#log.head()
def create_source_type(row):
    if row['traffic_source'] in ['yandex', 'google']:
        return 'organic'
    elif row['traffic_source'] in ['paid', 'email']:
        if row['region'] == 'Russia':
            return 'ad'
        return 'other'
    return row['traffic_source']
```

```python
log['source_type'] = log.apply(create_source_type, axis = 1)
log.head(10)
```

**Этап 4**

Откройте из папки ml-latest-small файл ratings.csv с оценками фильмов.
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
Примечание: под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
ratings['rating_count'] = ratings.groupby('userId')['userId'].transform('count')
#ratings
active_users = ratings[ratings['rating_count']>100].sort_values(by=['rating_count'])
#active_users
user_lifetime = active_users.groupby('userId')['timestamp'].agg(['min','max'])
user_lifetime['lifetime'] = user_lifetime['max'] - user_lifetime['min']
average_user_lifetime = user_lifetime['lifetime'].mean()/24/3600
print(f'Среднее время жизни пользователей, которые поставили > 100 оценок составляет {round(average_user_lifetime)} дня')
```

Откройте файл Python_13_join.ipynb, содержащий статистику услуг перевозки клиентов компании по типам.
Создайте 2 таблицы:
таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;
таблицу с тремя типами выручки для каждого client_id с указанием адреса клиента.
Примечание: в процессе объединения таблиц данные не должны потеряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
table = rzd.merge(auto, how = 'outer', on = 'client_id').merge(air, how = 'outer', on = 'client_id').fillna(0)
table
```

```python
table_with_addr = (
    client_base.merge(rzd, how='outer', on='client_id')
               .merge(auto, how='outer', on='client_id')
               .merge(air, how='outer', on='client_id')
               .fillna(0)
)
table_with_addr
```
