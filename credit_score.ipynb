```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Credit Card Defaulter Prediction.csv')
df
```

```python
#Gредварительный анализ данных. Видим, что есть данные с типом object. Выведим также уникальные значения
print(df.info())
print('-'*20)
print(f'Уникальные элементы SEX: {df["SEX"].unique()}')
print(f'\nУникальные элементы EDUCATION: {df["EDUCATION"].unique()}')
print(f'\nУникальные элементы MARRIAGE: {df["MARRIAGE"].unique()}')
#также видим, что null значения отсутсвуют
```

```python
#закодируем object в int
df1 = df.copy()
label_encoder = LabelEncoder()
df1['SEX'] = label_encoder.fit_transform(df1['SEX'])
df1['EDUCATION'] = label_encoder.fit_transform(df1['EDUCATION'])
df1['MARRIAGE'] = label_encoder.fit_transform(df1['MARRIAGE'])
```

```python
print(f'Расшифровки SEX: {df["SEX"].unique()}:{df1["SEX"].unique()}')
print(f'\nРасшифровки EDUCATION: {df["EDUCATION"].unique()}: {df1["EDUCATION"].unique()}')
print(f'\nРасшифровки MARRIAGE: {df["MARRIAGE"].unique()}: {df1["MARRIAGE"].unique()}')
```

```python
import numpy as np
df1['default '].unique()
```

```python
df1.rename(columns={'default ': 'DEFAULT'}, inplace=True)
df1.columns
```

```python
#обучение модели

import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')

models = {
    'Логистическая регрессия': LogisticRegression(),
    'Метод опорных векторов': SVC(),
    'Дерево решений': DecisionTreeClassifier(),
    'Случайный лес': RandomForestClassifier()
}
x_columns = df1.drop(columns= 'DEFAULT', axis=1)
y_columns = df1['DEFAULT']

X_train, X_test, y_train, y_test = train_test_split(x_columns, y_columns, test_size=0.2, random_state=42)

for model_name, model in models.items():
    print(f'{model_name}:')
    
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)
    
    report = classification_report(y_test, predictions)
    print(report)
 
        
```

Выводы
1. Логистическая регрессия и метод опорных векторов показали высокую точность (accuracy) для класса N, но они полностью не обнаружили класс Y. Это может быть связано с несбалансированностью классов и недостаточной способностью моделей обнаруживать редкий класс.

2. Дерево решений обладает хорошей точностью для класса N и неплохой способностью обнаруживать класс Y. Однако, F1-мера для класса Y ниже, что указывает на недостаточную способность модели обнаруживать этот класс.

3. Случайный лес демонстрирует наилучшие результаты среди всех рассмотренных моделей. Он имеет хорошую точность и способность обнаруживать оба класса, хотя F1-мера для класса Y все равно ниже.
